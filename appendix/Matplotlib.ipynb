```python
x = np.linspace(0, 2, 100)
# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend.
```

```python
labels = ['Test 1', 'Test 2', 'Test 3', 'Test 4', 'Test 5', 'Test 6',]

group_one_price = [30, 40, 50, 50, 30, 20]
group_two_price = [10, 20, 30, 40, 50, 20]
group_one_tax = [1,2,1,2,1,1]
group_two_tax = [3,4,3,4,3,4]

width = 0.50
fig, ax = plt.subplots()

#create our stacked bar visualization
ax.bar(labels, group_one_price, width, align = 'edge', yerr=group_one_tax, label='First Label')
ax.bar(labels, group_two_price, width, align = 'edge', yerr=group_two_tax, bottom=group_one_price, label='Second Label')

ax.set_ylabel('Value')
ax.set_title('Scores by Group')
ax.legend()
plt.show()
```

```python
plt.style.use('ggplot')
labels = ['Test 1', 'Test 2', 'Test 3', 'Test 4', 'Test 5', 'Test 6',]

group_one_price = [30, 40, 50, 50, 30, 20]
group_two_price = [10, 20, 30, 40, 50, 20]
group_one_tax = [1,2,1,2,1,1]
group_two_tax = [3,4,3,4,3,4]

width = 0.60
fig, ax = plt.subplots()

#create our stacked bar visualization
ax.bar(labels, group_one_price, width, align = 'edge', yerr=group_one_tax, label='First Label')
ax.bar(labels, group_two_price, width, align = 'edge', yerr=group_two_tax, bottom=group_one_price, label='Second Label')

ax.set_ylabel('Value')
ax.set_title('Scores by Group')
plt.xticks(rotation=90)
ax.legend()
plt.show()
```

```python
plt.style.use('ggplot')

data = {'Chocolate':20, "Vanilla":10, 'Coffee': 25, 'Pistachio': 15}
names = list(data.keys())
values = list(data.values())

fig, axs = plt.subplots(1,4, figsize=(5,4), sharey=True)

axs[0].bar(names,values)
axs[1].scatter(names,values)
axs[2].plot(names,values)
fig.suptitle('Plotting')
plt.xticks(rotation=90)
```
