```python
import numpy as np
import pandas as pd

sns.set_theme(style='whitegrid')

random_state = np.random.RandomState(365)
values = random_state.randn(365, 4).cumsum(axis=0)
dates = pd.date_range('2 2 2020', periods=365, freq='D')
df = pd.DataFrame(values, dates, columns=['A', 'B', 'C', 'D'])
df = df.rolling(7).mean()
```

```python
df.fillna(1)
```

```python
sns.set(rc={'figure.figsize':(15.37, 12.8)})
sns.lineplot(data=df, palette='tab10', linewidth=3.5)
```

```python
df_flights = pd.read_csv('data/flights.csv')
```

```python
#df_flights.head(10)
df_flights['DEPARTURE_DELAY'].isnull().values.any()
test_flights = df_flights.fillna(1)
#distance, air_time, flight_number, departure_delay
```

```python
#df_flights.dtypes 
```

```python
sns.set_theme(style="whitegrid")
cmap = sns.cubehelix_palette(rot=-.3, as_cmap=True)
g = sns.relplot(data=test_flights, x='DISTANCE', y='AIR_TIME', hue='FLIGHT_NUMBER', size='DEPARTURE_DELAY', palette=cmap)
g.fig.set_size_inches(15,15)
g.set(xscale='log', yscale='log')
g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
```
